```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('D5data3.csv')
```

```python
df.info()
```

```python
df.describe()
```

```python
x = df.iloc[:,:-1].values
```

```python
y = df.iloc[:,-1].values
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=1/3,random_state=0)
```

```python
# Linear Regression
from sklearn.linear_model import LinearRegression
LinMod = LinearRegression()

LinMod.fit(x_train,y_train)
```

```python
plt.scatter(x_train,y_train,c='r')
plt.plot(x_train,LinMod.predict(x_train))
```

```python
# Polynomial Regression
from sklearn.preprocessing import PolynomialFeatures
x_poly = PolynomialFeatures(degree=2)
x_poly = x_poly.fit_transform(x_train)
 
```

```python
# Fitting into model
PolyMod = LinearRegression()
```

```python
PolyMod.fit(x_poly,y_train)
```

```python
plt.scatter(x_train,y_train,c='orange')
plt.plot(x_train,PolyMod.predict(x_poly))
```
